```python
import torch.nn as nn
import torch.nn.functional as F
import torch.nn.init as init
class Autoencoder_Layer(nn.Module):
    def __init__(self):
        super(Autoencoder_Layer, self).__init__()
        self.id = nn.Embedding(1, int(hidden_size/2))
        self.Linear = nn.Linear(int(hidden_size/2), int(hidden_size/2))
        self.decoder = nn.Linear(int(hidden_size/2), hidden_size) # 수정: Linear 레이어 입력 크기 수정

        # 가중치 초기화
        for m in self.modules():
            if isinstance(m, nn.Linear) or isinstance(m, nn.Embedding):
                init.normal_(m.weight, mean=0, std=0.01)
                if hasattr(m, 'bias') and m.bias is not None:
                    init.constant_(m.bias, 0)
                    
    def forward(self,id):
        embed_user = self.id(id).view(-1)
        embed_user =  F.relu(self.Linear(embed_user))
        embed_user =  F.relu(self.Linear(embed_user))
        embed_user =  F.relu(self.Linear(embed_user))
        embed_user =  F.relu(self.Linear(embed_user))
        embed_user =  F.relu(self.Linear(embed_user))
        out = self.decoder(embed_user)
        return out
```
