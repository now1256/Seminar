```python
# 모델 softmax를 이용한 다중 분류 
class model(torch.nn.Module):
    def __init__(self):
        super(model,self).__init__()
        self.embedding_layer = nn.Embedding(num_embeddings=768, embedding_dim=3)
        self.linear1 = nn.Linear(11, 11)  
        self.linear2 = nn.Linear(11, 10)
        self.ReLU= nn.ReLU()
    def forward(self,a,b):
        b = self.embedding_layer(b)
        concat = torch.cat((a, b), dim=-1)
        x = self.ReLU(self.linear1(concat))
        output = self.linear2(x)
        return output
```

```python
classifier = model()
```

```python
# optimizer 경사하강법 loss_function cross entropy
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(classifier.parameters(), lr=0.1)
epochs = 1000
```

```python
for epoch in range(epochs):
    avg_cost = 0
    outputs = classifier(x_train1,x_train2)
    loss = loss_function(outputs,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    avg_cost += loss 
    print('Epoch:', '%04d' % (epoch + 1), 'cost =', '{:.9f}'.format(avg_cost))        
```
